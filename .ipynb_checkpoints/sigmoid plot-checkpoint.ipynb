#### This is a script used to show Sigmoid Function shape, which is generally used in machine learning field
$$ \sigma(z)=\frac {1}{1+e^{-z}}$$

The following figure shows a single neuron in network, x1~x3 is going to be the input, $\sigma(z)$ will be defined within this neuron
<img src="files\NeuronNetworkFigures\SingleNeuron.PNG" height="200" width="200">

If we have different weights for different inputs, and also set up a bias, then z is gonna be
$$ z = w \cdot x + b = \sum_j{w_jx_j} + b$$
Then let's take a look at shape of Sigmoid function:

```python
import numpy as np
import matplotlib.pyplot as plt
left_limit = -8.0
right_limit =8.0
z= np.arange(left_limit,right_limit+0.1,0.1)
sigma=1.0/(1.0+np.exp(-z))

plt.plot(z, sigma, 'r-')
plt.axis([left_limit,right_limit, -0.1, 1.25])
plt.show()
```

#### From the plot we could tell there are at least these advantages of using this function:
1. $ 0<\sigma(z)<1$ , which ganrantee it's a well-defined probability function
2. $\frac{\sigma(z)+\sigma(-z)}{2}=\frac{1}{2}$, which tells us this function is symmetric with respect to straight line y=0.5


Next, let's take a look at the shape of its derivative:
$$ \sigma(z)=\frac {1}{(1+e^{-z})(1+e^{z})}$$

```python
sigma_p=1.0/((1.0+np.exp(-z))*(1.0+np.exp(z)))
plt.plot(z, sigma_p, 'b-')
plt.axis([left_limit,right_limit,0.0,0.3])
plt.show()
```

```python
from math import cos

def recur_cos(value, r_times):
    if r_times == 1:
        return cos(value)
    else:
        return cos(recur_cos(value, r_times-1))
```

```python
x = []
y = []
for i in range(30,1,-1):
    x.append(i)
    y.append(recur_cos(0.5, i))
    
plt.plot(x,y)
plt.show()
```
